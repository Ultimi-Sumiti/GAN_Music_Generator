## Setup

```python
# Define the dataset.
DATA_FILE_PATH = "data/preprocessed/maestro-v3.0.0/dataset2/rolling_stones.h5"
dataset = MaestroV3DataSet(DATA_FILE_PATH, mode="pair")

# Load model from checkpoint.
ckp_path = "./lightning_logs/version_773/checkpoints/fa_classifier_epoch=15.ckpt"
model = GAN.load_from_checkpoint(ckp_path)
```

## Generate melody

```python
for i in range(10):
    print(i)
    melody_pm = generate_melody(model, n_bars=8, dataset=dataset, verbose=True)
    out_fn = "./outputs/songs/out_" + str(i) + ".midi"
    melody_pm.write(out_fn)
```
